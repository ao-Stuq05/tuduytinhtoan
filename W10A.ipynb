```python
def sum_power(n:int):
    total = 0
    for i in str(n):
        total += pow(int(i),2)
    return total

try:
    n = int(input())
except ValueError:
    print("n khong la so")
else:
    if n == 1: print("YES")
    else:
        my_list=[]
        my_list.append(n)
        while True:
            if sum_power(my_list[-1]) ==1:
                print("YES")
                break
            elif sum_power(my_list[-1]) in my_list:
                print("NO")
                break
            my_list.append(sum_power(my_list[-1]))
```

```python
num_arr = list(map(int,input().split()))

n = num_arr[0]
num_arr.pop(0)

try:
    for i in range(n-1):
        for j in range(i+1,n):
            assert num_arr[i] != num_arr[j], "Contain duplicate"
except:
    print("mang khong hop le")
else:
    total = 0
    for i in num_arr:
        total += i
    print(total)
```

```python
import re

S = input("Nhap string S: ")
T = int(input("Nhap so lan testcase: "))
pattern_arr = []
for i in range(T):
    pattern_arr.append(input("Nhap testcase: "))

for pattern in pattern_arr:
    if re.match(pattern,S):
        print("Valid")
    else:
        print("Invalid")
```
