```python
#Viết một chương trình nhập vào một số nguyên n và in ra các chữ số đảo ngược nhau.
n = int(input("Nhập số n: "))
m = 0
while n > 0:
    m = m * 10 + n%10
    n = n // 10
print(m)
```

```python
#Hoán đổi hai số không sử dụng biến tạm thời. Sử dụng phép toán XOR bitwise để hoán đổi giá trị của hai biến.
a, b = map(int, input("Nhập hai số a, b: ").split())
a = a ^ b
b = a ^ b
a = a ^ b
print(a, b)
```

```python
#Kiểm tra xem một số có phải là lũy thừa của 2 hay không. Sử dụng các toán tử bitwise.
n = int(input("Nhập số n: "))
if n & (n-1) == 0:
    print(n, "là lũy thừa của 2")
else:
    print(n, "không phải là lũy thừa của 2")
```

```python
#Nhập vào 3 số nguyên dương a, b, c là độ dài 3 cạnh của một tam giác. Kiểm tra 3 số đã nhập có tạo thành một tam giác hợp lệ hay không. Nếu không in ra "Không phải tam giác".
#Nếu có, phân loại tam giác đó là "Tam giác đều", "Tam giác cân", hay "Tam giác thường".
a, b, c = map(int, input().split())
if a <= 0 or b <= 0 or c <= 0: print("Không phải tam giác")
else:
  if a < b+c and b < a+c and c < a+b:
    if a == b and b == c: print("Tam giác đều")
    elif a == b or b == c or c == a: print("Tam giác cân")
    else: print("Tam giác thường")
  else: print("Không phải tam giác")
```

```python
#Nhập vào một số nguyên dương n là số năm, kiểm tra xem năm đó có phải là năm nhuận hay không. Năm nhuận là năm chia hết cho 4 nhưng không chia hết cho 100, hoặc chia hết cho 400.
#Nếu là năm nhuận in ra "Yes", nếu không in ra "No".
n = int(input())
if n < 0: print("No")
else:
  if (n%4 == 0 and n%100 != 0) or n%400 == 0: print("Yes")
  else: print("No")
```

```python
#Nhập vào số kWh điện tiêu thụ. Tính tiền điện theo các bậc sau:
# 0-50 kWh: 1500đ/kWh
# 51-100 kWh: 2000đ/kWh
# 100 kWh: 3000đ/kWh
tien_dien = int(input())
if (tien_dien >= 0 and tien_dien <= 50): print("Tiền điện tiêu thụ: ", tien_dien*1500, "đ")
elif (51 <= tien_dien and tien_dien <= 100): print("Tiền điện tiêu thụ: ", tien_dien*2000, "đ")
else: print("Tiền điện tiêu thụ: ", tien_dien*3000, "đ")
```

```python
#Nhập vào hai số thực a và b của phương trình ax + b = 0. Tìm và in ra nghiệm của phương trình, kết quả làm tròn đến số thập phân thứ 2. In ra "Vô nghiệm" hoặc "Vô số nghiệm" trong trường hợp tương ứng.
a = float(input())
b = float(input())
if a == 0:
  if b == 0: print("Vô số nghiệm")
  else: print("Vô nghiệm")
else: print(round(-b/a, 2))
```

```python
#Nhập vào điểm trung bình của một học sinh và in ra học lực của học sinh đó. Xếp loại học lực theo quy tắc:
# >= 8.0: Giỏi
# >= 6.5: Khá
# >= 5.0: Trung bình
# < 5.0: Yếu
diem = float(input())
if diem >= 8.0: print("Giỏi")
elif 6.5 <= diem and diem < 8.0: print("Khá")
elif 5.0 <= diem and diem < 6.5: print("Trung bình")
else: print("Yếu")
```
