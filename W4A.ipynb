```python
#W4A10
n = int(input())
def collatz_length(x):
    length = 1
    while x != 1:
        if x % 2 == 0 :
            x = x // 2
        else:
            x = 3 * x + 1
        length += 1
    return length
max_length = 0
res = 1
for i in range (1, n + 1):
    length = collatz_length(i)
    if length > max_length  or (length == max_length and i < res):
        max_length = length
        res = i
print (res , max_length)
```
