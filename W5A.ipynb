```python
#W5A1
def Max_Of_Two(a, b):
  if (a > b): return a
  else: return b

a, b = map(int, input().split())

print(Max_Of_Two(a, b))
```

```python
#W5A2
def sw(a, b):
  return b, a

a, b = map(int, input().split())
print(sw(a, b))
```

```python
#W5A3
from math import sqrt
def prime_number_check(n):
  if (n == 1): return False
  elif (n == 2): return True
  else:
    for (i) in range (2, int(sqrt(n)) + 1):
      if (n%i == 0): return False
    return True

n = int(input())

print(prime_number_check(n))
```

```python
#W5A4
def perfect_number_check(n):
  sum = 0
  for (i) in range (1, n):
    if (n%i == 0): sum += i
  if (sum == n): return True
  else: return False

n = int(input())

print(perfect_number_check(n))
```

```python
#W5A5
def find_k_in_list(lst, k):
  for (i) in range (0, len(lst)):
    if (lst[i] == k): return i+1
  return -1

lst = list(map(int, input().split()))
k = int(input())

print(find_k_in_list(lst, k))
```

```python
#W5A6
def factorial(n):
  product = 1
  for (i) in range (1, n+1): product *= i
  return product

n = int(input())

print(factorial(n))
```

```python
def he(s):
    if s == "+": return 1
    elif s == "-": return 2
    elif s == "*": return 3
    return 4

def calculator(aa):
    li = aa.split(" ")
    arr = []
    arr.append(float(li[0]))
    dau = -1
    for i in range(1, len(li)):
        if i % 2 == 1:
            dau = he(li[i])
        else:
            a = float(li[i])
            if dau == 3:
                b = arr[len(arr) - 1]
                arr.pop()
                a *= b
                arr.append(a)
            elif dau == 4:
                b = arr[len(arr) - 1]
                arr.pop()
                b = b / a
                arr.append(b)
            elif dau == 1:
                arr.append(a)
            else: arr.append(-a)
    ans = 0
    for i in arr:
        ans += i
    return float(f"{ans:.2f}")
```

```python
def Hamming(x, y):
    x = x ^ y
    ans = 0
    while x > 0:
        ans += x % 2
        x = int(x / 2)
    return ans
```

```python
#W5A9
def sum_of_numbers(n):
  sum = 0
  while(n > 0):
    sum += n%10
    n //= 10
  return sum

n = int(input())

print(sum_of_numbers(n))
```

```python
#W5A10
def dangcau(s1, s2):
    mp1 = {}
    mp2 = {}
    for i in range(len(s1)):
        if s1[i] in mp1:
            if mp1[s1[i]] != s2[i]: return "false"
        else: mp1[s1[i]] = s2[i]
        if s2[i] in mp2:
            if mp2[s2[i]] != s1[i]: return "false"
        else: mp2[s2[i]] = s1[i]
    return "true"

s1, s2 = input().split()
print(dangcau(s2, s2))
```
