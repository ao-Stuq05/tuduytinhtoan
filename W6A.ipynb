```python
#W6A1
lst = list(map(int, input().split()))
seen = set()
new_lst = []

for item in lst:
    if item not in seen:
        new_lst.append(item)
        seen.add(item)
lst = new_lst

print(lst)
```

```python
import ast

#W6A9
# Example input: {'a': 1}
dic1_str = input()
# Example input: {'a': 2}
dic2_str = input()

dic1 = ast.literal_eval(dic1_str)
dic2 = ast.literal_eval(dic2_str)
dic3 = {}

# Merge dic1 into dic3
for key, value in dic1.items():
  dic3[key] = dic3.get(key, 0) + value

# Merge dic2 into dic3
for key, value in dic2.items():
  dic3[key] = dic3.get(key, 0) + value

print(dic3)
```

```python
#W6A10
a = list(map(int, input().split()))
k = int(input())

res = []

n = len(a)
for i in range(n):
    for j in range(i + 1, n):
        if a[i] + a[j] == k:
            res.append((a[i], a[j]))

res.sort(key=lambda x: x[0])

print(res)
```

```python
#W6A12
a = list(map(int, input().split()))

cnt = {}
for x in a:
    cnt[x] = cnt.get(x, 0) + 1

max_freq = max(cnt.values())

res = min(x for x in cnt if cnt[x] == max_freq)

print(res)
```

```python
#W6A19
def max_of_(lst):
  max = lst[0]
  m = 0
  for (i) in range(1, len(lst)):
    if (lst[i] > max):
      max = lst[i]
      m = i
  return m +1

lst = list(map(int, input().split()))

print(max_of_(lst))
```

```python
#W6A20
def find_k_in(lst, k):
  for (i) in range(len(lst)):
    if (lst[i] == k):
      return i +1
  return -1

lst = list(map(int, input().split()))
k = int(input())

print(find_k_in(lst, k))
```
