```python
#W8A1
class Cylinder:
  def __init__(self, radius, height):
    self.radius = radius
    self.height = height
  def total_area(self):
    return 2 * 3.14 * self.radius * self.height
  def volume(self):
    return 3.14 * self.radius**2 * self.height

r, h = map(int, input().split())
t = Cylinder(r, h)

print(t.total_area())
print(t.volume())
```

```python
#W8A2
class Date:
  def __init__(self, day, month, year):
    self.day = day
    self.month = month
    self.year = year

  def _is_leap_year(self, year):
    return (year % 4 == 0 and year % 100 != 0) or (year % 400 == 0)

  def date_check(self):
    is_leap = self._is_leap_year(self.year)
    if self.month in [1, 3, 5, 7, 8, 10, 12]:
      return 0 < self.day <= 31
    elif self.month in [4, 6, 9, 11]:
      return 0 < self.day <= 30
    elif self.month == 2:
      max_feb_day = 29 if is_leap else 28
      return 0 < self.day <= max_feb_day
    else: # Invalid month
      return False

  def next_day(self):
    is_leap = self._is_leap_year(self.year)

    if self.month in [1, 3, 5, 7, 8, 10]: # Months with 31 days, excluding December
      if self.day == 31:
        self.day = 1
        self.month += 1
      else:
        self.day += 1
    elif self.month == 12: # December
      if self.day == 31:
        self.day = 1
        self.month = 1
        self.year += 1
      else:
        self.day += 1
    elif self.month in [4, 6, 9, 11]: # Months with 30 days
      if self.day == 30:
        self.day = 1
        self.month += 1
      else:
        self.day += 1
    elif self.month == 2: # February
      max_feb_day = 29 if is_leap else 28
      if self.day == max_feb_day:
        self.day = 1
        self.month += 1
      else:
        self.day += 1

d, m, y = map(int, input().split())
t = Date(d, m, y)
if t.date_check():
  t.next_day()
  print(f"{t.day} {t.month} {t.year}") # Added print to show the next day
else: print("INVALID")
```

```python
#W8A3
class Point:
  def __init__(self, diaphragm, pitch):
    self.dia = diaphragm
    self.pit = pitch
  def pitch_check(self):
    if self.dia == 0 and self.pit != 0: return True
    return False
  def O_check(self):
    if self.dia == self.pit == 0: return True
    return False
  def distance(self):
    print(f"{(self.dia**2 + self.pit**2)**0.5:.2f}")

x, y = map(int, input().split())
m = Point(x, y)
if m.O_check():
  print('Yes, O. ')
  m.distance()
elif m.pitch_check():
  print('Yes, pitch. ')
  m.distance()
else:
  print('No.')
  m.distance()
```

```python
#W8A5
class ShoppingCart:
    def __init__(self):
        self.cart = []

    def add_product(self, name, price):
        self.cart.append({"name": name, "price": price})

    def remove_product(self, name):
        for item in self.cart:
            if item["name"] == name:
                self.cart.remove(item)
                return "Đã xóa sản phẩm"
        return "Không tìm thấy sản phẩm"

    def is_empty(self):
        return len(self.cart) == 0

    def total_price(self):
        total = 0
        for item in self.cart:
            total += item["price"]
        return total

    def show_products(self):
        if self.is_empty():
            print("Giỏ hàng rỗng")
        else:
            for item in self.cart:
                print(item["name"], "-", item["price"])

    def clear_cart(self):
        self.cart.clear()


# ===== Chương trình chính =====
cart = ShoppingCart()

while True:
    print("\n===== MENU =====")
    print("2. Thêm sản phẩm")
    print("3. Xóa sản phẩm theo tên")
    print("4. Kiểm tra giỏ hàng rỗng")
    print("5. Tính tổng tiền")
    print("6. Hiển thị sản phẩm")
    print("7. Xóa toàn bộ giỏ hàng")

    choice = int(input("Chọn chức năng (2-7): "))

    if choice == 2:
        name = input("Nhập tên sản phẩm: ")
        price = float(input("Nhập giá: "))
        cart.add_product(name, price)

    elif choice == 3:
        name = input("Nhập tên sản phẩm cần xóa: ")
        print(cart.remove_product(name))

    elif choice == 4:
        print("Giỏ hàng rỗng:", cart.is_empty())

    elif choice == 5:
        print("Tổng tiền:", cart.total_price())

    elif choice == 6:
        cart.show_products()

    elif choice == 7:
        cart.clear_cart()
        print("Đã xóa toàn bộ giỏ hàng")

    else:
        print("Lựa chọn không hợp lệ")

    exit_program = input("Có muốn thoát khỏi chương trình không? ")
    if exit_program.lower() == "yes":
        break
```

```python
#W8A6
class Rectangle:
    def __init__(self, w, h):
        self.w = w
        self.h = h

    def area(self):
        return self.w * self.h

    def perimeter(self):
        return 2 * (self.w + self.h)

    def scale(self, k):
        self.w *= k
        self.h *= k


# ===== Chương trình chính =====
w, h, k = map(float, input().split())

rect = Rectangle(w, h)
rect.scale(k)

print(f"{rect.area():.2f} {rect.perimeter():.2f}")
```

```python
#W8A7
class BankAccount:
    def __init__(self, owner, balance):
        self.owner = owner
        self.balance = float(balance)

    def deposit(self, amount):
        if amount > 0:
            self.balance += amount

    def withdraw(self, amount):
        # Nếu rút quá số dư thì bỏ qua giao dịch
        if 0 < amount <= self.balance:
            self.balance -= amount

    def __str__(self):
        return f"{self.owner}: {self.balance:.2f}"

# Phần chạy chương trình
try:
    # Đọc dòng đầu tiên: owner và balance
    first_line = input().split()
    owner = first_line[0]
    initial_balance = float(first_line[1])

    account = BankAccount(owner, initial_balance)

    # Đọc số lượng lệnh
    n = int(input())

    for _ in range(n):
        line = input().split()
        cmd = line[0]
        amount = float(line[1])

        if cmd == 'DEPOSIT':
            account.deposit(amount)
        elif cmd == 'WITHDRAW':
            account.withdraw(amount)

    print(f"{account.balance:.2f}")

except Exception as e:
    print("Lỗi nhập liệu:", e)
```

```python
#W8A8
class Student:
    def __init__(self, name):
        self.name = name
        self.scores = []

    def add_score(self, subject, score):
        self.scores.append(score)

    def avg(self):
        if not self.scores:
            return 0.0
        return sum(self.scores) / len(self.scores)

    def rank(self):
        average = self.avg()
        if average >= 8.0:
            return "Excellent"
        elif average >= 6.5:
            return "Good"
        elif average >= 5.0:
            return "Average"
        else:
            return "Poor"

# Phần chạy chương trình
try:
    name = input().strip()
    n = int(input())

    student = Student(name)

    for _ in range(n):
        # Input dạng: Subject Score (ví dụ: Math 9)
        parts = input().split()
        # Điểm là phần tử cuối cùng
        score = float(parts[-1])
        # Tên môn học là phần còn lại (không dùng trong tính toán nhưng cần tách ra)
        subject = " ".join(parts[:-1])

        student.add_score(subject, score)

    average_score = student.avg()
    ranking = student.rank()

    print(f"{student.name} {average_score:.2f} {ranking}")

except Exception as e:
    print("Lỗi:", e)
```

```python
#W8A9
class Book:
    def __init__(self, title, author, year):
        self.title = title
        self.author = author
        self.year = int(year)

class Library:
    def __init__(self):
        self.books = []

    def add(self, book):
        self.books.append(book)

    def count_by_author(self, author_name):
        count = 0
        for book in self.books:
            # So sánh chính xác tên tác giả
            if book.author == author_name:
                count += 1
        return count

    def find_by_year(self, year):
        count = 0
        for book in self.books:
            if book.year == year:
                count += 1
        return count

# Phần chạy chương trình
try:
    n = int(input())
    lib = Library()

    for _ in range(n):
        line = input().strip()

        if line.startswith("ADD"):
            # Cấu trúc: ADD title; author; year
            # Cắt bỏ chữ "ADD " đầu tiên (4 ký tự)
            content = line[4:]
            parts = content.split(';')
            title = parts[0].strip()
            author = parts[1].strip()
            year = int(parts[2].strip())
            lib.add(Book(title, author, year))

        elif line.startswith("COUNTYEAR"):
            # Cấu trúc: COUNTYEAR year
            year_query = int(line.split()[1])
            print(lib.find_by_year(year_query))

        elif line.startswith("COUNT"):
            # Cấu trúc: COUNT author
            # Tên tác giả có thể có dấu cách, lấy phần sau chữ "COUNT "
            author_query = line[6:].strip()
            print(lib.count_by_author(author_query))

except Exception as e:
    print("Lỗi:", e)
```

```python
#W8A10

import math

class Point3D:
    def __init__(self, x, y, z):
        self.x = x
        self.y = y
        self.z = z

def subtract(p1, p2):
    return Point3D(p1.x - p2.x, p1.y - p2.y, p1.z - p2.z)

def cross_product(v1, v2):
    return Point3D(
        v1.y * v2.z - v1.z * v2.y,
        v1.z * v2.x - v1.x * v2.z,
        v1.x * v2.y - v1.y * v2.x
    )

def dot_product(v1, v2):
    return v1.x * v2.x + v1.y * v2.y + v1.z * v2.z

def magnitude(v):
    return math.sqrt(v.x**2 + v.y**2 + v.z**2)

try:
    points = []
    # Đọc 4 điểm A, B, C, D
    for _ in range(4):
        coords = list(map(float, input().split()))
        points.append(Point3D(coords[0], coords[1], coords[2]))

    A, B, C, D = points[0], points[1], points[2], points[3]

    # Mặt phẳng (A, B, C): Vectơ pháp tuyến n1 = AB x AC
    AB = subtract(B, A)
    AC = subtract(C, A)
    n1 = cross_product(AB, AC)

    # Mặt phẳng (B, C, D): Vectơ pháp tuyến n2 = BC x BD
    BC = subtract(C, B)
    BD = subtract(D, B)
    n2 = cross_product(BC, BD)

    # Tính cos góc
    dot = dot_product(n1, n2)
    mag1 = magnitude(n1)
    mag2 = magnitude(n2)

    # Tránh chia cho 0
    if mag1 == 0 or mag2 == 0:
        print("0.00")
    else:
        cos_phi = abs(dot) / (mag1 * mag2)
        # Đảm bảo giá trị trong [-1, 1] do sai số float
        cos_phi = max(min(cos_phi, 1), -1)

        angle_rad = math.acos(cos_phi)
        angle_deg = math.degrees(angle_rad)

        print(f"{angle_deg:.2f}")

except Exception as e:
    print(e)
```

```python
#W8A11
class ComplexNumber:
    def __init__(self, real, imag):
        self.real = real
        self.imag = imag

    def multiply(self, other):
        new_real = self.real * other.real - self.imag * other.imag
        new_imag = self.real * other.imag + self.imag * other.real
        return ComplexNumber(new_real, new_imag)

try:
    # Nhập số phức 1
    line1 = list(map(float, input().split()))
    c1 = ComplexNumber(line1[0], line1[1])

    # Nhập số phức 2
    line2 = list(map(float, input().split()))
    c2 = ComplexNumber(line2[0], line2[1])

    result = c1.multiply(c2)

    # In ra: phần thực phần ảo
    # Sử dụng format g để xóa số 0 thừa nếu cần, hoặc .2f tùy đề
    # Theo test case mẫu: -13 82 -> in số nguyên nếu tròn, số thực nếu lẻ
    # Tuy nhiên để đơn giản và khớp với đa số bài chấm tự động float:
    print(f"{int(result.real) if result.real.is_integer() else result.real} {int(result.imag) if result.imag.is_integer() else result.imag}")

except:
    print("Lỗi nhập liệu")
```

```python
#W8A12
import math

class Triangle:
    def __init__(self, e1, e2, e3):
        self.e1 = e1
        self.e2 = e2
        self.e3 = e3

    def get_area(self):
        a, b, c = self.e1, self.e2, self.e3
        # Kiểm tra điều kiện tam giác: tổng 2 cạnh > cạnh còn lại
        if (a + b > c) and (a + c > b) and (b + c > a):
            p = (a + b + c) / 2
            area = math.sqrt(p * (p - a) * (p - b) * (p - c))
            return area
        else:
            return -1

try:
    # Nhập 3 cạnh trên 1 dòng
    edges = list(map(int, input().split()))
    t = Triangle(edges[0], edges[1], edges[2])

    area = t.get_area()

    if area == -1:
        print("invalid")
    else:
        print(f"{area:.2f}")

except:
    print("invalid")
```
